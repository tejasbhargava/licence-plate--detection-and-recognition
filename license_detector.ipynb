```python
from ultralytics import YOLO
```

```python
import cv2
```

```python
import os

os.environ["KMP_DUPLICATE_LIB_OK"] = "TRUE"
os.environ["OMP_NUM_THREADS"] = "1"
os.environ["MKL_NUM_THREADS"] = "1"

print("Env vars set")
```

```python
model = YOLO("yolov8n.pt")
print("YOLO loaded")
```

```python
my_model = YOLO("models/best.pt")
```

```python
my_model(
    source="carLicence1.mp4",
    conf=0.4,
    save=True,   # saves result
    show=False   # IMPORTANT
)
```

```python
metrics = my_model.val(
    data="data/data.yaml",
    imgsz=640
)
```

```python
precision = metrics.box.p[0]
recall = metrics.box.r[0]
f1 = metrics.box.f1[0]
map50 = metrics.box.map50
map5095 = metrics.box.map
```

```python
print("\nYOLOv8 Detection Metrics")
print("------------------------")
print(f"Precision    : {precision:.3f}")
print(f"Recall       : {recall:.3f}")
print(f"F1 Score     : {f1:.3f}")
print(f"mAP@0.5      : {map50:.3f}")
print(f"mAP@0.5:0.95 : {map5095:.3f}")
```
